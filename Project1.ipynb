#Data Cleaning
In this step, we will:

Check and remove duplicate rows.
Handle missing values.
Fix data types (if necessary).
Clean text data (optional).
Verify that the dataset is clean.

```python
## Create a Copy of the Dataset
df_clean = df.copy()
#This keeps the original dataset safe in case you need to compare it later.
```

```python
## check duplicate rows
df_clean.duplicated().sum()
```

```python
##Remove Duplicate Rows
df_clean.drop_duplicates(inplace=True)
df_clean.duplicated().sum()
```

```python
## check missing values
df_clean.isnull().sum()
```

```python
## Handle missing values
## fill missing values with unknown
df_clean['director'].fillna('unknown',inplace = True)
df_clean['cast'].fillna('unknown',inplace = True)
df_clean['country'].fillna('unknown',inplace = True)

## remove rows where important information is missing
df_clean.dropna(subset = ['date_added','rating','duration'],inplace=True)
```

```python
# check missing values
df_clean.isnull().sum()
# now we can see misssing values are zero
```

```python
# check datatypes
df_clean.dtypes
```

```python
print(df_clean['date_added'].dtype)
df_clean.info()
```

```python
## save cleaned dataset
df_clean.to_csv("netflix_cleaned.csv", index=False)
```

"The dataset was cleaned by removing duplicate records and handling missing values. Missing values in the director, cast, and country columns were replaced with 'Unknown' to preserve records, while rows with missing values in critical columns (date_added, rating, and duration) were removed. The date_added column was converted to datetime format to facilitate date-based analysis. Finally, the cleaned dataset was verified and saved for further analysis."

#Exploratory Data Analysis (EDA)


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Import Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns


# Improve plot appearance
sns.set_style("whitegrid")

# Display plots with larger size
plt.rcParams["figure.figsize"] = (10,6)
```

```python
## How many Movies and TV Shows are available on Netflix?
# count Movies and Tv shows
count = df_clean['type'].value_counts()
print(count)
```

```python
##Visualization
plt.figure(figsize=(6,5))
sns.countplot(x = 'type',data = df_clean)
plt.title("Distribution of Movies and TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

```python
###Which Countries Produce the Most Netflix Content?

top_country = df_clean['country'].value_counts().head(10)

print(top_country)
```

```python
### Which Release Years Have the Most Netflix Content?
release_year_count = df_clean['release_year'].value_counts().sort_index()

print(release_year_count)
```

```python
### Visualization
plt.figure(figsize=(14,6))

sns.lineplot(
    x=release_year_count.index,
    y=release_year_count.values,
    marker='o'
)

plt.title("Netflix Content Released by Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
### Which Content Ratings Are Most Common?
rating_count = df_clean['rating'].value_counts()

print(rating_count)
```

```python
##Visualization

plt.figure(figsize=(10,6))

sns.countplot(
    y='rating',
    data=df_clean,
    order=df_clean['rating'].value_counts().index
)

plt.title("Distribution of Netflix Content Ratings")
plt.xlabel("Number of Titles")
plt.ylabel("Rating")

plt.show()
```

```python
###Which Genres Are Most Popular?
genres = (
    df_clean['listed_in']
    .str.split(', ')
    .explode()
)

top_genres = genres.value_counts().head(10)

print(top_genres)
```

```python
##visualization
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_genres.values,
    y=top_genres.index
)

plt.title("Top 10 Most Popular Netflix Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")

plt.show()
```

```python
##when Netflix added content to its platform.
# Convert 'date_added' to datetime objects, coercing errors to NaT
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'], errors='coerce')

# Drop rows where 'date_added' became NaT after conversion
df_clean.dropna(subset=['date_added'], inplace=True)

content_added = df_clean['date_added'].dt.year.value_counts().sort_index()

print(content_added)
```

```python
## ##visualization
plt.figure(figsize =(12,6))
sns.lineplot(x = content_added.index,y = content_added.values,marker ='o')

plt.title('content added to netflix by year')
plt.xlabel('year')
plt.ylabel('number of Titles added')
```

#Visualizations

```python
#Chart 1 – – Histogram
plt.figure(figsize=(10,6))

plt.hist(df_clean['release_year'], bins=30, edgecolor='black')

plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Frequency")

plt.show()
```

```python
###Scatter Plot
plt.figure(figsize=(10,6))

plt.scatter(
    df_clean['release_year'],
    df_clean['duration_num'],
    alpha=0.5
)

plt.title("Release Year vs Duration")
plt.xlabel("Release Year")
plt.ylabel("Duration")

plt.show()
```
